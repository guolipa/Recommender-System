````python
'''
#方式一
fig,([ax1,ax2],[ax3,ax4]) = plt.subplots(2,2)
#方式二
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
'''
fig,ax = plt.subplots(2,2)  #ax可以像二维数组一样索引
ax[0,0].hist(np.random.randn(100),bins = 20, color = 'k', alpha = 0.3)
ax[0,1].scatter(np.arange(30),np.arange(30) + 3 * np.random.randn(30))
ax[1,0].plot(np.random.randn(50).cumsum(),'k--',color = 'r')

plt.show()
````

### 颜色、标记、线类型

````python
plt.figure()
plt.plot(np.random.randn(30),'ko--')
plt.plot(np.random.randn(30),linestyle = '--',marker = 'o', color = 'g')
plt.show()
````

### 刻度、标签、图例和注释

````python
fig,ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum(),'k',color = 'k',label = 'a')
ax.plot(np.random.randn(1000).cumsum(),'k--',color = 'g',label = 'b')
ax.plot(np.random.randn(1000).cumsum(),'k.',color = 'y',label = 'c')
ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(['one','two','three','four','five'],rotation = 30, fontsize = 'small')
ax.set_xlabel('stage')
ax.set_ylabel('values')
ax.set_title('values of different stages')
ax.legend(loc = 'best')
plt.show()
````

````python
fig, ax = plt.subplots()
x = np.arange(0., 6., 1)
ax.plot(x,x,'r',x,x**2,'g',x, x**3,'b',x, np.exp(x),'k',linewidth = 2.5) #一个函数中可以绘制多条曲线
ax.plot(x,x,'rx',x,x**2,'go',x, x**3,'bs',x, np.exp(x),'kD',linewidth = 2.5)
ax.legend()
plt.show()
````

#### 2.2 利用种类变量

````python
names = ['a', 'b', 'c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
````

````python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
plt.show()
````

## 2. Pandas

###  plot

````python
ser = pd.Series(np.random.randn(10).cumsum(), index = np.arange(0,100,10))  #Series类型
df = pd.DataFrame(np.random.randn(10,4).cumsum(0),columns = ['A','B','C','D'],index = np.arange(0,100,10))#DataFrame类型

fig, axs = plt.subplots(2,1)
ser.plot(ax = axs[0],style='k--',kind = 'line',xticks = range(0,100,5))
df.plot(ax = axs[1],)
plt.show()
````

### plot.bar() & plot.barh()

````python
fig, axs = plt.subplots(2,1)

data = pd.Series(np.random.rand(12),index = list('abcdefghesfo'))

data.plot.bar(ax = axs[0],color = 'k',alpha = 0.7, figsize = (8,8))
data.plot.barh(ax = axs[1],color = 'g',alpha = 0.7, figsize = (10,8))

plt.show()
````

````python
#DataFrame
df = pd.DataFrame(np.random.rand(6,4),index = ['one','two','three','four','five','six'],columns = list('ABCD'))

fig, axs = plt.subplots(3,1)
df.plot.bar(ax=axs[0],figsize = (8,10))
df.plot.barh(ax=axs[1],figsize = (8,15))
df.plot.barh(ax=axs[2],stacked = True)
plt.show()
````
