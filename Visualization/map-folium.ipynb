在地图中添加标记部件Marker

添加坐标符号标记

- location：同folium.Map()中的同名参数，用于确定标记部件的经纬位置

- popup：str型或folium.Popup()对象输入，用于控制标记部件的具体样式（folium内部自建了许多样式），默认为None，即不显示部件

- icon：folium.Icon()对象，用于设置popup定义的部件的具体颜色、图标内容等

```python
folium.Marker([38.521050, -95.828807],popup='<i>Mt. Hood Meadows</i>',icon=folium.Icon(color='blue')).add_to(map)
map
```

```python
folium.Marker([35.521050, -93.828807],popup='Timberline Lodge',icon=folium.Icon(color='green')).add_to(map)
map
```

添加圆圈

- location：同folium.Map()中的location，用于控制圆圈的圆心坐标

- radius：int型，用于控制圆圈的半径，单位米，注意，在folium.Circle()中，radius因为半径的单位是米，所以其大小随着我们对地图的缩放程度而进行相应的变化，但在与folium.CircleMarker()方法中的radius参数单位为像素，即其为屏幕上大小固定的一个圆圈，随着地图的缩放，其大小也不会发生变化

- color：str型，用于控制圆圈的颜色，默认为十六进制颜色'#3388ff'，即一种蓝色

- fill：bool型，当为True时，圆圈内部将被填充上色彩，默认不填充

- fill_color：str型，控制圆圈内部填充的色彩，默认与color参数一致

- fill_opacity：float型，用于控制圆圈内部填充颜色的透明度，从0.到1.之间，默认为0.2

- popup：str型或folium.Popup()对象，用于控制圆圈的样式，默认为None，即无样式

```python
m = folium.Map(location = [38.321050, -95.828807],tiles = 'Mapbox Bright',zoom_start=5)  
#tooltip = 'Click me!'
#folium.Marker([38.621050, -94.828807], popup='<b>Timberline Lodge</b>', tooltip=tooltip).add_to(m)
folium.Circle(
    radius=100,
    location=[38.321050, -95.828807],
    popup='The Waterfront', #圆圈的样式
    #color='crimson',
    fill=True,
    fill_opacity = 1
).add_to(m)

folium.Circle(
    radius=150,
    location=[39.32210, -94.838807],
    #popup='The Waterfront',
    color='crimson',
    fill=False,
    #fill_opacity = 1
).add_to(m)

m
```

### 热力图

```python
from folium import plugins
from folium.plugins import HeatMap


map_hooray = folium.Map(location=[38.321050, -95.828807],
                    zoom_start = 6) 

# Ensure you're handing it floats
loc['latitude'] = loc['latitude'].astype(float)
loc['longitude'] = loc['longitude'].astype(float)

heat_df = loc[['latitude', 'longitude']]
heat_df = heat_df.dropna(axis=0, subset=['latitude','longitude'])

# List comprehension to make out list of lists
heat_data = [[row['latitude'],row['longitude']] for index, row in heat_df.iterrows()]

# Plot it on the map
HeatMap(heat_data).add_to(map_hooray)

# Display the map
map_hooray.save(r'D:\software\Github\Recommender-System\RS-LBSN\dataset\Foursquare\f\heatmap.html')
map_hooray
```

```python
nyc_venues = nyc.loc[nyc.city_v == 'New York'][['latitude','longtitude']]
```

```python
f, ax = plt.subplots(1, 1, sharey=True,figsize=(5,6))

lon_min, lon_max = -74.075,-73.8
lat_min, lat_max = 40.625,40.88
nyc_venues = nyc_venues[(nyc_venues["longtitude"]>lon_min) &\
                    (nyc_venues["longtitude"]<lon_max) &\
                    (nyc_venues["latitude"]>lat_min) &\
                    (nyc_venues["latitude"]<lat_max)]
nyc_venues.plot(kind='scatter', x='longtitude', y='latitude',
                color='yellow', 
                s=.09, alpha=.6, subplots=True, ax=ax)
ax.set_title("New York City")
ax.set_facecolor('black')

plt.show()
```
