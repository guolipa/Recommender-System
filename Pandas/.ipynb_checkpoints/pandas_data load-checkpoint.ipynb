````python
import pandas as pd
````

```read_csv()  参数：```

- basic
    - filepath：文件路径
    - sep : str， defaults to ',' for read_csv(), \t for read_table()
- Column and Index Locations and Names  
    - header : int or list of ints, default 'infer'. 
    - names : array-like, default None，给列命名，列名。

**可以让pandas自动分配列名，也可以自己指定列名。自动分配列名时，设定header = None。**

````python
pd.read_csv('example.csv',header = None) #自动分配列名
````

````python
pd.read_csv('example.csv',names = ['id','name','age','country','text']) #自定义列名
````

**可以将某一列作为对象的行索引**

````python
names = ['id','name','age','country','text']
pd.read_csv('example.csv',names = names, index_col = 'id')
````

**有些情况下一张表的分隔符不是固定的，比如使用空格或其他方式来分隔字段，当字段中分隔符不固定时，需要函数中传入一个正则表达式re来表示分隔符**

````python
pd.read_table('example.txt',sep = '\s+',names = names, index_col = 'id')
````

** 参数skiprows = []可以指定跳过读取行数**
