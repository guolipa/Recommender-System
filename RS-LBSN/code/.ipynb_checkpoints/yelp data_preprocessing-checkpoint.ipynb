```python
state = business['state']
```

```python
state_count = state.value_counts()   #统计各州中POI的数目
state_count[state_count >10]   
```

```python
state_uniq = state.unique()  #统计州的数目
state_uniq
```

```python
city = business['city']
```

```python
city_count = city.value_counts() #统计各城市中POI的数目
city_count[city_count > 1000]
```

```python
city_uniq = city.unique()  #统计城市的数目
city_uniq
```

```python
#将POI按照州和城市索引划分
state_index = business['state']
city_index = business['city']
business.index = [state_index, city_index]
business = business[['business_id','latitude', 'longitude', 'stars', 'review_count','categories']].sort_index(level = 0)
```
