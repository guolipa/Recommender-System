## 1.HITS算法构建

user - venue提取

```python
user_venue = nyc[['uid','vid']]
m = len(user_venue.uid.unique()) #用户数目
n = len(user_venue.vid.unique()) #地点数目
user_venue = user_venue.values  #用户签到数据
matrix = np.zeros((m,n),dtype = np.int32) #user-venue数组
for row in user_venue:
    matrix[row[0]][row[1]] = 1   
matrix
```

```python
将hub 和 authority的初始值设置为次数
#uv= nyc[['uid','vid']]
#con1 = uv.uid.value_counts()
#h =con1.sort_index().values.astype(np.int32)
#con2 = uv.vid.value_counts()
#a =con2.sort_index().values.astype(np.int32)
```

```python
'''
input:array/matrix
output:
    
'''
def HIST(matrix,max_iterations,delta):
    m = matrix.shape[0] #用户数
    n = matrix.shape[1] #地点数
    hub = np.ones(m)  #存储用户的hub值,初始化为1
    authority = np.ones(n)  #存储地点的autyority值，初始化为1
    flag = False  
    c = []
    for iter in range(max_iterations):
        change = 0.0  # 记录每轮的变化值
        norm = 0  # 标准化系数
        temp = {}
        #计算每个地点的authority
        temp = authority.copy()
        authority = np.dot(hub,matrix)
        norm = sum(pow(authority,2))
        norm = np.sqrt(norm)
        authority /= norm
        change += sum(abs(temp - authority))
        #计算每个用户的hub
        norm = 0
        temp = hub.copy()
        hub = np.dot(authority,matrix.T)
        norm = sum(pow(hub,2))
        norm = np.sqrt(norm)
        hub /= norm
        change += sum(abs(temp - hub))
        c.append(change)
        if change < delta:
            flag = True
            break;
    if flag:
        print("finished in %s iterations!" % (iter + 1))
    else:
        print("finished out of %d iterations!" % max_iterations)
    return hub, authority,c
```

```python
hub, authority ,c= HIST(matrix, 100, 0.0001)
```

## 2. NYC种类对应的地点

```python
#利用下方的user_category函数
def catsplit(x):
    if x != None:
        return str(x).split(',')
    else:
        return None
'''
input: x,DataFrame,having 'uid,category'
output: user-category-counts
'''
def venue_category(x):
    cat = []  #整理用户种类
    pref = x[['vid','category']]
    pref['category'] = pref['category'].map(catsplit)
    for i in range(len(pref)):
        n,c= pref.iloc[i]
        if c != None:
            for p in c:
                cat.append([n,p])
    cpref = pd.DataFrame(cat,columns = ['vid','cid'])
    #cpref.index = cpref['uid']
    cpref['cid'] = cpref['cid'].replace('nan',np.nan)
    cpref = cpref.dropna()
    cpref.index = np.arange(len(cpref))
    cpref['nums'] = np.ones(len(cpref))
    vc = cpref.groupby(['vid','cid'],as_index = False).count()
    print('Done')
    return vc
```

训练集和测试集划分

```python
'''
输入参数：
    总体的数据集:out，DataFrame格式
    划分比例：ratio，测试集占的比例
返回参数：
    训练集：train， DataFrame格式
    测试集：test
'''
def train_test_split(out,ratio):    
    train = out.iloc[0:2,:]
    test = out.iloc[0:2,:] #初始化训练集和测试集
    #获得用户uid集合
    users = out.uid.unique()

    for user in users:
        tips = out.loc[out.uid == user] #某个用户的所有签到
        tips_NYC = tips.loc[tips.city_v == 'New York'] #该用户在NYC的所有签到
        tips_nNYC = tips.loc[tips.city_v != 'New York'] ##该用户不在NYC的所有签到
        num = len(tips)  #用户总的签到数目
        num_NYC = len(tips_NYC) #用户在NYC的签到数目
        if num < 20:
            l = 4
        else:
            l = int(len(tips) * ratio)  #该用户需要加入集的80%的签到数的数目
        if num_NYC > l:  #
            test = pd.concat([test,tips_NYC.iloc[0:l,:]],axis = 0)
            train = pd.concat([train,tips_NYC.iloc[l:,:]],axis = 0)
            train = pd.concat([train,tips_nNYC],axis = 0)
        else:
            test = pd.concat([test,tips_NYC],axis = 0)
            train = pd.concat([train,tips_nNYC],axis = 0)
    train = train.iloc[2:,:]
    test = test.iloc[2:,:]
    print('用户的tips总数：%d' % len(out))
    print('训练集的tips总数：%d' % len(train))
    print('测试集的tips总数：%d' % len(test))
    print(len(train)/ len(out))
    return train, test
```

```python
out_train,out_test = train_test_split(out,0.2)
home_train, home_test = train_test_split(home,0.2)
```

```python
def catsplit(x):
    if x != None:
        return str(x).split(',')
    else:
        return None
'''
input: x,DataFrame,having 'uid,category'
output: user-category-counts
'''
def user_category(x):
    cat = []  #整理用户种类
    pref = x[['uid','category']]
    pref['category'] = pref['category'].map(catsplit)
    for i in range(len(pref)):
        n,c= pref.iloc[i]
        if c != None:
            for p in c:
                cat.append([n,p])
    cpref = pd.DataFrame(cat,columns = ['uid','cid'])
    #cpref.index = cpref['uid']
    cpref['cid'] = cpref['cid'].replace('nan',np.nan)
    cpref = cpref.dropna()
    cpref.index = np.arange(len(cpref))
    cpref['nums'] = np.ones(len(cpref))
    uc = cpref.groupby(['uid','cid'],as_index = False).count()
    print('Done')
    return uc
```

```python
#种类评分划分
#将用户访问种类的次数划分为1-5分
def score_map(x):
    if x == 1:
        return 2
    elif x == 2:
        return 3
    elif x == 3 or x == 4:
        return 4
    elif x >= 5:
        return 5
    else:
        return 1
```

```python
#cv
#基于种类偏好的候选集
def candidate(train, test,cnum):
    candidatelist = {} #所有用户的候选列表
    userSet = test.uid.unique() #用户集合
    for uid in userSet:
        train_venues = train[train.uid == uid].vid.unique() #用户在训练集中的地点
        categoryRating = preference.iloc[uid].sort_values(ascending = False) #该用户的所有种类评分
        category = categoryRating.iloc[0:cnum].index  #选择cnum个种类
        can = []
        for cid in category:
            if cid in cv.keys():
                venues = cv[cid] #种类对应的地点
                can.extend(venues)
        candidatelist[uid] = np.setdiff1d(can,train_venues)
    return candidatelist
```

```python
def popularity(candidatelist,topN):
    recommendlist = {}
    for uid in candidatelist.keys():
        venue_pop = pops.loc[candidatelist[uid]].sort_values(by = 'vid', ascending = False) #用户候选集中地点的流行度
        venue_topN = venue_pop.iloc[0:topN].index
        recommendlist[uid] = venue_topN
    return recommendlist    
```

```python
def most_popularity(train,test, topN):
    all_venues = pops.index
    recommendlist = {}
    userSet = test.uid.unique()
    for uid in userSet:
        train_venues = train[train.uid == uid].vid.unique() #用户在训练集中的地点
        recommend_venues = np.setdiff1d(all_venues,train_venues)
        recommendlist[uid] = pops.loc[recommend_venues].sort_values(by = 'vid', ascending = False).iloc[0:topN].index
    return recommendlist
```

```python
def hits(candidatelist,topN):
    recommendlist = {}
    for uid in candidatelist.keys():
        venue_authority = authority.loc[candidatelist[uid]].sort_values(ascending = False) #用户候选集中地点的流行度
        venue_topN = venue_authority.iloc[0:topN].index
        recommendlist[uid] = venue_topN
    return recommendlist   
```

```python
def UserSimilarity(data):
    #建立item 和 user的倒排表
    item_user = dict()
    for row in data:
        user = row[0]
        item = row[1]
        if item not in item_user:
            item_user[item] = set()
        item_user[item].add(user)
    #计算用户间的共同items
    C = dict();N = dict()
    for item, users in item_user.items():
        for u in users:
            N[u] += 1
            for v in users:
                if u == v:
                    continue
                C[u][v] += 1
    #计算用户相似矩阵
    W = dict()
    for u, related_users in C.items():
        for v, cuv in related_users.items():
            W[u][v] = cuv / math.sqrt(N[u] * N[v])
    return W
```

```python
def recommend(candidatelist,topN,method):
    if method == 'popularity':
        return popularity(candidatelist,topN)
    if method == 'hits':
        return hits(candidatelist,topN)
    
```

```python
def evaluate(test, recommendlist,N):
    hit = 0
    n_recall = 0
    n_precision = 0
    userSet = test.uid.unique() #用户集合
    for uid in userSet:
        test_venues = test.loc[test.uid == uid].vid.unique()
        recommend_venues = np.array(recommendlist[uid])
        inter = np.intersect1d(test_venues,recommend_venues)
        hit += len(inter)
        n_recall += len(test_venues)
        n_precision += N
    precision = hit / (1.0 * n_precision)
    recall = hit / (1.0 * n_recall)
    return precision, recall
```
